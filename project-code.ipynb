```python
# For each topic segment, extract top keywords using TF-IDF:
# TF-IDF ranks words by how frequent and distinctive they are in a group of sentences.
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict

# Group sentences by cluster label
clustered_sentences = defaultdict(list)
for sent, label in zip(sentences, cluster_labels):
    clustered_sentences[label].append(sent)

def extract_keywords(text_list, top_n=3):
    """Extracts top N keywords using TF-IDF from a list of sentences."""
    vectorizer = TfidfVectorizer(stop_words='english', max_features=20)
    X = vectorizer.fit_transform(text_list)
    sums = X.sum(axis=0)
    terms = vectorizer.get_feature_names_out()
    term_weights = [(terms[i], sums[0, i]) for i in range(len(terms))]
    ranked_terms = sorted(term_weights, key=lambda x: x[1], reverse=True)
    return [term for term, _ in ranked_terms[:top_n]]

segment_keywords = {}
for label, sents in clustered_sentences.items():
    segment_keywords[label] = extract_keywords(sents)
    print(f"Segment {label}: Keywords - {segment_keywords[label]}")
    print("Sample sentences:", sents[:3])
    print("-"*50)
```

```python
# Parameters (assume each sentence 10 seconds)
sentence_duration_sec = 10
block_length_sec = 60  # 1 minute segments

start_times = np.arange(0, len(sentences)*sentence_duration_sec, sentence_duration_sec)

segments = []
current_seg_sents = []
current_start_idx = 0

for i in range(len(sentences)):
    current_seg_sents.append(sentences[i])
    
    # When reached 1 minute block or end of sentences, form segment and caption
    if ((i - current_start_idx + 1) * sentence_duration_sec >= block_length_sec) or (i == len(sentences) - 1):
        caption_text = " ".join(current_seg_sents)  # Concatenate sentences for caption
        
        segment = {
            'start': start_times[current_start_idx],
            'end': start_times[i] + sentence_duration_sec,
            'caption': caption_text,
            'sentences': current_seg_sents
        }
        segments.append(segment)
        
        # Reset for next segment
        current_seg_sents = []
        current_start_idx = i + 1

# Output captions and timestamps
for i, seg in enumerate(segments):
    start_m, start_s = seg['start'] // 60, seg['start'] % 60
    end_m, end_s = seg['end'] // 60, seg['end'] % 60
    
    print(f"Caption {i}: {start_m:02}m {start_s:02}s - {end_m:02}m {end_s:02}s")
    print("Caption Text:", seg['caption'])
    print("-" * 50)
```

```python
import spacy
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.text import TfidfVectorizer

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Your transcript text
transcript_text = "Your raw transcript text from video goes here."

# Sentence segmentation, filtering empty sentences
doc = nlp(transcript_text)
sentences = [sent.text.strip() for sent in doc.sents if sent.text.strip()]

print(f"Number of sentences after segmentation: {len(sentences)}")

if len(sentences) < 2:
    # Handle short transcript
    print("Not enough sentences to cluster. Processing as a single segment.")
    # Encode the only sentence or entire transcript
    model = SentenceTransformer('all-MiniLM-L6-v2')
    if len(sentences) == 1:
        embeddings = model.encode(sentences[0])
        print(f"Single sentence embedding: {embeddings}")
        # You can choose to display or process this single sentence
    else:
        # For empty or no sentences
        print("No valid sentences found.")
else:
    # Proceed with clustering if enough sentences
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(sentences)

    n_clusters = min(5, len(sentences))
    cluster_model = AgglomerativeClustering(n_clusters=n_clusters)
    clusters = cluster_model.fit_predict(embeddings)

    # Extract keywords for each cluster
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(sentences)
    feature_names = vectorizer.get_feature_names_out()

    keywords_per_cluster = []
    for cluster_id in set(clusters):
        cluster_sentences = [sentences[i] for i in range(len(sentences)) if clusters[i] == cluster_id]
        tfidf_cluster = vectorizer.transform(cluster_sentences)
        scores = tfidf_cluster.sum(axis=0).A1
        top_indices = scores.argsort()[-5:][::-1]
        keywords = [feature_names[idx] for idx in top_indices]
        keywords_per_cluster.append((cluster_id, keywords))

    print("Clusters and their keywords:")
    for cid, kws in keywords_per_cluster:
        print(f"Cluster {cid}: {kws}")
```

```python
import spacy
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython.display import YouTubeVideo, display
import numpy as np

# Step 1: Display YouTube video
video_id = "N8svLoC2eNA"
display(YouTubeVideo(video_id, width=640, height=360))

# Step 2: Sample transcript text 
transcript_text = """
Artificial Intelligence and Machine Learning are transforming the world.
In this course, we will cover the fundamentals of AI and ML algorithms.
From supervised learning to deep learning, you will learn how machines can mimic human intelligence.
We will also explore real-world applications and data-driven insights.
"""

# Step 3: Sentence segmentation
nlp = spacy.load("en_core_web_sm")
doc = nlp(transcript_text)
sentences = [sent.text.strip() for sent in doc.sents if sent.text.strip()]

# Example video duration and segment length in seconds
total_duration = 60  # e.g., 1-minute video for demo
segment_length = 10  # 10 seconds per segment
num_segments = int(np.ceil(total_duration / segment_length))

# Step 4: Approximate number of sentences per segment
sent_per_segment = max(1, int(np.ceil(len(sentences) / num_segments)))

# Group sentences by segments
segments = [
    " ".join(sentences[i * sent_per_segment : (i + 1) * sent_per_segment])
    for i in range(num_segments)
]

# Step 5: Extract one keyword per segment via TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=5)
tfidf_matrix = vectorizer.fit_transform(segments)
feature_names = vectorizer.get_feature_names_out()

# Pick top keyword per segment
keywords_per_segment = []
for i in range(len(segments)):
    row = tfidf_matrix.getrow(i).toarray().flatten()
    if row.sum() == 0:
        keywords_per_segment.append("N/A")  # no keyword found
    else:
        top_index = row.argmax()
        keywords_per_segment.append(feature_names[top_index])

# Display time, segment text, and keyword label
for idx, (seg_text, kw) in enumerate(zip(segments, keywords_per_segment)):
    start_time = idx * segment_length
    end_time = min(total_duration, (idx + 1) * segment_length)
    print(f"Segment {idx+1}: Time {start_time}-{end_time} sec")
    print(f"Text: {seg_text}")
    print(f"Topic Keyword: {kw}\n")
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

nlp = spacy.load("en_core_web_sm")

def process_video(video_id, segment_length=10):
    # Fetch transcript with timestamps
    transcript = YouTubeTranscriptApi.get_transcript(video_id)

    # Combine transcript parts into time-aligned segments
    # Aggregate transcript text within non-overlapping segment_length windows
    total_duration = transcript[-1]['start'] + transcript[-1]['duration']
    num_segments = int(np.ceil(total_duration / segment_length))
    
    segments_text = [''] * num_segments
    
    for entry in transcript:
        segment_index = int(entry['start'] // segment_length)
        segments_text[segment_index] += ' ' + entry['text']

    # Extract keywords per segment
    vectorizer = TfidfVectorizer(stop_words='english', max_features=5)
    tfidf_matrix = vectorizer.fit_transform(segments_text)
    feature_names = vectorizer.get_feature_names_out()

    keywords_per_segment = []
    for i in range(num_segments):
        row = tfidf_matrix.getrow(i).toarray().flatten()
        if row.sum() == 0:
            keywords_per_segment.append("N/A")
        else:
            top_index = row.argmax()
            keywords_per_segment.append(feature_names[top_index])

    # Print or return segments with timing and keywords
    for idx, (text, kw) in enumerate(zip(segments_text, keywords_per_segment)):
        start_time = idx * segment_length
        end_time = min(total_duration, (idx + 1) * segment_length)
        print(f"Video: {video_id} | Segment {idx+1}: Time {start_time:.2f}-{end_time:.2f} sec")
        print(f"Text: {text.strip()}")
        print(f"Keyword: {kw}\n")

# Example usage on first video
process_video(video_id)
```
